```python
import pandas as pd
import numpy as np

data = pd.read_csv("./data/agaricus-lepiota.data")
data
```

```python
for col in data.columns:
    print(f"{col}: {np.unique(data[col])}")
```

```python
ms_col = []
for col in data.columns:
    value_col = np.unique(data[[col]])
    for value in value_col:
        if value == "?":
            ms_col.append(col)
            
ms_col
```

```python
encode_data = data

decode_ms = {}

for col in ms_col:
    values, counts = np.unique(encode_data[col], return_counts=True)
    decode_ms[col] = {}
    for value, count in zip(values, counts):
        decode_ms[col][value] = count 
```

```python
# Convert data to numeric value
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for col in encode_data.columns:
    encode_data[col] = label_encoder.fit_transform(encode_data[col])

for col in encode_data.columns:
    print(f"{col}: {np.unique(encode_data[col])}, type: {encode_data[col].dtype}")
    
```

```python
encode_ms = {}

for col in ms_col:
    values, counts = np.unique(encode_data[col], return_counts=True)
    encode_ms[col] = {}
    for value, count in zip(values, counts):
        encode_ms[col][value] = count 
```

```python
print(encode_ms)
print(decode_ms)
```

```python
# Match encode_ms and decode_ms we will see LableEncoder convert missing values to different values
for col in ms_col:
    values = np.unique(encode_data[col])

    avg = int(sum(values)/(len(values))) 
    print(f"{col}: avg = {avg}")
    encode_data[col].replace(0, avg, inplace=True)

ms_col
```

```python
# if we use RandomUnderSample, the dataset size will decrease but if we use SMOTE then the dataset size will increase. So that we use both of them together. 
# RandomUnderSample will remove samples from the majority class
# SMOTe will add samples to the minority class
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
import pandas as pd

# Original column names
resample_col = encode_data.columns

# Apply RandomUnderSampler with a higher sampling ratio
rus = RandomUnderSampler(sampling_strategy='auto')  
X_resampled, y_resampled = rus.fit_resample(encode_data.drop('p', axis=1), encode_data['p'])

# Apply SMOTE with a slightly lower ratio
smote = SMOTE(sampling_strategy='auto')  
X_final, y_final = smote.fit_resample(X_resampled, y_resampled)

# Re-create DataFrame with the final balanced dataset
encode_data = pd.DataFrame(
    data=pd.concat([y_final, X_final], axis=1),
    columns=resample_col
)
data
```
